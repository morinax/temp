```python
allfiles = glob.glob("./e-stat_data/data*")

dfs = []
for file in allfiles:
    try:
        #print(file)
        temp = pd.read_excel(file, header=9)
        #print(temp.columns)
        file_year = file.split('_')[2]
        file_year = file_year.split('.')[0]
        temp = temp.rename(columns={'品   目   分   類':'品目分類'})
        temp = temp[['品目分類', 'Unnamed: 11']]
        temp = temp.rename(columns={'Unnamed: 11':'ポリ袋・ラップ（円/世帯)'})
        temp = temp[temp['品目分類']=='ポリ袋・ラップ']
        temp['年'] = file_year
        temp['世帯'] = "二人以上"
        temp = temp[['世帯', '年', 'ポリ袋・ラップ（円/世帯)']]
    except:
        #print(file)
        temp = pd.read_excel(file, header=9, engine='openpyxl')
        temp = temp.rename(columns={'品   目   分   類':'品目分類'})
        #print(temp.columns)
        file_year = file.split('_')[2]
        file_year = file_year.split('.')[0]
        temp = temp[['品目分類', '複数単位']]
        temp = temp.rename(columns={'複数単位':'ポリ袋・ラップ（円/世帯)'})
        temp = temp[temp['品目分類']=='ポリ袋・ラップ']
        temp['年'] = file_year
        temp['世帯'] = "二人以上"
        temp = temp[['世帯', '年', 'ポリ袋・ラップ（円/世帯)']]
    
    dfs.append(temp)
    #display(temp.head(5))
df = pd.concat(dfs)
df.head()
```
